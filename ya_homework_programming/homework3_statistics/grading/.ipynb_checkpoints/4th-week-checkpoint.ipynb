```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
# В рамках выполнения задания используется распределение хи-квадрат
# Сокращенная ссылка на вики - https://goo.gl/Zo6VuR

#  В рамках задания используется распределение хи-квадрат с 3-мя степенями свободы
# генерация данных выборки и данных оси x
chi2 = sts.chi2(3)
sample = chi2.rvs(1000)
x =  np.linspace(0,10,1000)
```

```python
# график гистограммы случайной выборки
plt.hist(sample, normed=True, bins=20, label='histogramm')

# график теоретической плотности распределения случайной величины 
pdf = chi2.pdf(x)
plt.plot(x, pdf, label='theoretical pdf')
plt.legend(loc='best')
plt.ylabel('$f(x)$'+'\n'+'fraction of samples')
plt.xlabel('$x$')
```

```python
# n - размеры выборок
n_list = [2,3,5,7,10,15,20,30]
fig, ax = plt.subplots(nrows=len(n_list), figsize=(10, len(n_list)*10))

# для распределения хи-квадрат Математическое ожидание равно степениям свободы, в данном случае - 3,
# дисперсия равняется степени свободы умноженной на два.
print('Расчеты:\n' + 'Количество степеней свободы - 3 \n')

for ind, n in enumerate(n_list):
    norm = sts.norm(3, np.sqrt(6/n))
#   вывод выкладок с вычислениями параметров нормальных распределений, аппроксимирующих выборочные средние при различных n
    print('Для n (Выборка) равное ' + str(n) + '\n'\
          + 'Математическое ожидание - 3' + '\n'\
          + 'Дисперсия - ' + 'sqrt(2*3/'+ str(n) + ')' + ' = ' + str(2*3/n) + '\n')
#   определяем ось x в рамках 3 сигм
    norm_x = np.linspace(3-3*np.sqrt(6/n),3+3*np.sqrt(6/n),100)
    norm_y = norm.pdf(norm_x)
    
    ax[ind].plot(norm_x, norm_y, label='theoretical pdf')
    
    averages = []
    for i in range(1000):
        m = np.random.choice(sample, n)
        averages.append(np.mean(m))
    ax[ind].hist(averages, normed=True, bins=20, label='histogramm')
    ax[ind].set_xlim(0,7)
    ax[ind].legend(loc='best')
    ax[ind].set_ylabel('$f(x)$')
    ax[ind].set_xlabel('$x$')
    ax[ind].set_title('n = '+str(n))
```

Несмотря на то, что распределение хи-квадрат ассиметрично, очевидное подтверждение центральной пределеной теоремы можно наблюдать уже при выборке равной 7 и даже меньших ее значениях. При этом считаю интересным отметить, что типичные для подобной функции выбросы справа практически нивелируются при n=20 и выше.
