```python
# This part generates Raylegh distribution and histogram for empirical estimate for it.
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.rayleigh.html
left = 0.0
right = 5.0

Rayleigh  = sts.rayleigh ()
x = np.linspace(left,right,100)
# theoretical density function
pdf = Rayleigh.pdf(x) 
# empirical sample of the function
sample = Rayleigh.rvs(1000) 
plt.plot(x, pdf, label='theoretical pdf') 
plt.hist(sample, normed=True, label='empirical histogram')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
# this part calculates normal distributuin parameters for all three n's and generates analytical estimates
# information about mean and variance for Railegh function can be found here
# https://en.wikipedia.org/wiki/Rayleigh_distribution, in our case sigma is 1.0.
import math
pdf = []
for i in range(len(N)):
    n = N[i]
    E = math.pi/2.
    D = (4. - math.pi/2.)/n
    Sigma = math.sqrt(D)
    norm_rv = sts.norm(E, Sigma)
    pdf.append(norm_rv.pdf(x))
```

```python
# plot comparison for N = 5
plt.plot(x, pdf[0], label="normal distributuion")
plt.hist(means[0], bins=30, normed=True, label="histogram")
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```
