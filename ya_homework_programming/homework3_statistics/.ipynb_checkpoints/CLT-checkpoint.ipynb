Выберем для исследования Бета (Beta) распределение.

Плотность распределения случайной величины дается формулой: $$
f(x;\alpha,\beta)  = \frac{\Gamma(\alpha+\beta)}{\Gamma(\alpha)\Gamma(\beta)}\, x^{\alpha-1}(1-x)^{\beta-1}.$$
Мат. ожидание: $$\operatorname{E}X = \frac{\alpha}{\alpha+\beta}$$
Дисперсия: $$\operatorname{D}(X) = \operatorname{E}[(X - \mu)^2] = \frac{\alpha \beta}{(\alpha + \beta)^2(\alpha + \beta + 1)}$$

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Построим график и гистограмму, считая, что $$\alpha = 3, \beta = 6$$

```python
fig, ax = plt.subplots(1, 1)
x = np.linspace(0,1,200)
# параметры исходного распределения
a = 3
b = 6

rv = sts.beta(a,b)
pdf = rv.pdf(x)
#вычисляем мат. ожидание и дисперсию исходного распределения
mean, var = sts.beta.stats(a, b, moments='mv')
print("Параметры распределения:")
print("Alpha = " + str(a)+ ", Beta = "+ str(b))
print("Мат.ожидание: "+ str(mean)+",  Дисперсия: "+ str(var))
print("Расчет по формулам (для проверки):")
print("Мат.ожидание: "+ str(float(a)/(a+b))+",  Дисперсия: "+str(float(a)*b/((a+b)**2*(a+b+1))))
plt.xlabel('x')
plt.ylabel('f(x)')
ax.plot(x, pdf, label="$a = 2, b = 6$")
# генерируем выборку из 1000 элементов
r = sts.beta.rvs(a, b, size=1000)
ax.hist(r, normed=True, histtype='stepfilled', alpha=0.5)
plt.legend()
plt.title("Beta distribution density.")
```

Генерируем по 1000 выборок содержащих 5, 10 и 50 элементов.
Считаем выборочные средние: $$\bar{X}=\frac{1}{n}\sum X_i.$$
Cтроим гистограммы с наложенными соответственными нормальными распределениями $$N(EX,\frac{DX}{n})$$, где $$EX$$ и $$DX$$ - мат. ожидание и дисперсия исходного распределения, а n  - число элементов в выборке.

```python
x = np.linspace(0,1,200)
for n in [5,10,50]: # перебираем размеры  выборок
    plt.figure()
    x_avg = []
    for i in range(0, 1000): # строим 1000 выборок
        # создаем список x_avg выборочных средних  
        x_avg.append( sum(sts.beta.rvs(a, b, size=n)) / float(n) )
    # строим гистограмму выборочных средних. 
    plt.subplot()
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.hist(x_avg, normed=True, histtype='stepfilled', alpha=0.5)
    plt.title("$n = %s$" % n)
    # строим соответствующее нормальное распределение
    print("Параметры нормального распределения:")
    print("Мат.ожидание: "+ str(mean)+",  Дисперсия: "+ str(var/n))
    plt.plot(x, sts.norm.pdf(x, mean, np.sqrt(var/n)),'r-', alpha=0.6)
    plt.show()
```

Из рисунков хорошо видно, что с ростом выборки вид гистограммы приближается к соответствующему нормальному распределению.
